```julia
include("fld.jl")
using PyPlot
using CSV
```

```julia
Nr = 5
Nz = 50
Lz = 0.2
println("Size of domain in Z is $(Lz)")
Lr = 50
dr = Lr/Nr
dz = Lz/Nz
zcenters = linspace(dz/2,Lz-dz/2,Nz)
T = ones(Nr*Nz)*0.001
rho = ones(Nr,Nz)
Er = a*(0.001*ones(Nr*Nz)).^4
sigma_func(t,T,Nr,Nz,Lr,Lz) = 300.0*(reshape(T,Nr,Nz).^-3)
D_func(t,T,Nr,Nz,Lr,Lz,Er,sigma) = 1.0./(3.0*sigma)
Cv_func(t,T,Nr,Nz,Lr,Lz) = 0.3*ones(Nr,Nz)
EOS(t,T,Nr,Nz,Lr,Lz) = 0.3*T
EOS(t,T,i,j,Nr,Nz,Lr,Lz) = 0.3*T
invEOS(t,E,Nr,Nz,Lr,Lz) = E/0.3
single_invEOS(t,E,i,j,Nr,Nz,Lr,Lz) = E/0.3

#println(Er,a*T.^4,EOS(0,T,Nr,Nz,Lr,Lz))
function Q_func(t,T,Nr,Nz,Lr,Lz)
    hr = Lr/Nr
    hz = Lz/Nz
    Q = zeros(Nr,Nz)
    Q
end
```

```julia
tfinal = 1.0
dt = 0.001
steps = convert(Int,round(tfinal/dt))
times, Er_t,T_t, diff_solves, outer_its = gray_GARNET(tfinal,dt,T,Er,D_func,sigma_func,Q_func,Cv_func,EOS,invEOS,single_invEOS,rho,
                              Nr,Nz,Lr,Lz,lower_z = a*0.5,upper_z="refl",upper_r="refl", LOUD=1, 
                              fname="tmp.jld", max_its = 2000, conserve = false)
```

```julia
st = steps+1
Tplot = reshape(T_t[:,st],Nr, Nz)
plot(zcenters,Tplot[3,:],"-",label="T")
Trplot = reshape((Er_t[:,st]/a).^.25,Nr, Nz)
plot(zcenters,Trplot[3,:],"-",label="Tr")
axis([0,0.2,0,1.1])
legend(loc="best")
#plot(zcenters,(Er_t[:,steps+1]/a).^.25,"-")
```
